```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,100)
k1 = 1
k2 = .75
k3 = .5
a1 = 0
a2 = 1.5
a3 = 2
b1 = 0
b2 = .5
b3 = .75

plt.xlabel('x')
plt.ylabel('y')

plt.plot(x, k1*np.cos(x - a1)+b1)
plt.plot(x, k2*np.cos(x - a2)+b2)
plt.plot(x, k3*np.cos(x - a3)+b3)
```
